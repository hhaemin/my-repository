### Numpy modules

```python
import numpy as np

arr = np.random.randn(100)
arr
```

#### Statistics modules
    - max, min, median, mean, std, var
    - cov, corrcoef

```python
np.max(arr), np.min(arr), np.mean(arr), np.var(arr), np.std(arr)
```

#### linspace, logspace
    - linspace : 설정한 범위에서 선형적으로 분할한 위치의 값을 출력
    - logspace : 설정한 범위에서 로그로 분할한 위치의 값을 출력

```python
np.linspace(0, 100, 5)
```

##### logspace
    - log10(x1)=2, log_10(x_2)=4

```python
np.logspace(2, 4, 3)
```

#### numpy random module
    - seed : 랜덤의 설정값
    - rand : 균등분포로 난수 생생
    - randn : 정규분포로 난수 생성
    - randint : 균등분포로 정수 생성
    - shuffle : 섞기
    - choice : 데이터를 특정 확률로 뽑기

##### seed

```python
np.random.seed(1)
res1 = np.random.randint(10, 100, 10)

np.random.seed(2)
res2 = np.random.randint(10, 100, 10)

np.random.seed(1)
res3 = np.random.randint(10, 100, 10)

res1, res2, res3
```
