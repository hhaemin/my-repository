```python
df = sns.load_dataset('titanic')
mean_age = df['age'].mean(axis=0)
mean_age
# 데이터 통계의 수치 확인
```

```python
df['age'].fillna(mean_age, inplace=True) # 누락값을 mean_age로 대체하며 실행 즉시 저장된다.
```

```python
most_freq = df['embark_town'].value_counts(dropna=True).idxmax() # 최빈 값을 저장한다.
most_freq
```

```python
df['embark_town'].value_counts(dropna=False)
```

```python
df['embark_town'].fillna(most_freq, inplace=True)
df['embark_town'].value_counts(dropna=False)
```
